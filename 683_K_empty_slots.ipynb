There is a garden with N slots. In each slot, there is a flower. The N flowers will bloom one by one in N days. In each day, there will be exactly one flower blooming and it will be in the status of blooming since then.

Given an array flowers consists of number from 1 to N. Each number in the array represents the place where the flower will open in that day.

For example, flowers[i] = x means that the unique flower that blooms at day i will be at position x, where i and x will be in the range from 1 to N.

Also given an integer k, you need to output in which day there exists two flowers in the status of blooming, and also the number of flowers between them is k and these flowers are not blooming.

If there isn't such day, output -1.

```python
#68s
class Solution(object):
    def kEmptySlots(self, flowers, k):
        """
        :type flowers: List[int]
        :type k: int
        :rtype: int
        """
        def find_interval(pos, possible_intervals):
            # Returns index of interval in possible_intervals that contains pos, inclusive.
            if pos < possible_intervals[0][0]:
                return -1
            l, r = 0, len(possible_intervals) - 1
            while l < r - 1:
                mid = l + (r - l) / 2
                if pos == possible_intervals[l][0]:
                    return l
                elif pos < possible_intervals[mid][0]:
                    # Go to left half
                    r = mid
                else: # pos >= possible_intervals[mid][0]:
                    l = mid
            if pos >= possible_intervals[r][0]:
                l = r
            if pos >= possible_intervals[l][0] and pos <= possible_intervals[l][1]:
                return l
            else:
                return -1

        if k >= len(flowers) - 1:
            return -1
        possible_intervals = [[0, len(flowers) - 1]] # Intervals of continuous empty positions. Ordered in terms of start position. 
        day = 0
        while len(possible_intervals) > 0:
            blossom_pos = flowers[day]-1 # Off by one
            index = find_interval(blossom_pos, possible_intervals) # Find interval that contains new blossom position.
            if index == -1: # The new blossom is outside interesting intervals and does not effect result.
                day += 1
                continue
            to_break = possible_intervals.pop(index)
            if to_break[1] - (blossom_pos + 1) + 1 == k and to_break[1] != len(flowers) - 1:
                return day + 1
            elif to_break[1] - (blossom_pos + 1) + 1 > k: # This new interval could break into a length-k interval in the future. 
                possible_intervals.insert(index, [blossom_pos + 1, to_break[1]])
            if (blossom_pos - 1) - to_break[0] + 1 == k and to_break[0] != 0:
                return day + 1
            elif (blossom_pos - 1) - to_break[0] + 1 > k: # This new interval could break into a length-k interval in the future. 
                possible_intervals.insert(index, [to_break[0], blossom_pos - 1])
            day += 1

        return -1
```

```python
class Solution(object):
    def kEmptySlots(self, flowers, k):
        """
        :type flowers: List[int]
        :type k: int
        :rtype: int
        """
        if k >= len(flowers)-1:
            return -1
        record = set()
        for i in range(len(flowers)):
            record.add(flowers[i])
            if flowers[i] - k - 1 >= 0 and flowers[i] - k - 1 in record:
                if self.check(record, flowers[i]-k-1, flowers[i]):
                    return i + 1
            if flowers[i] + k + 1 <= len(flowers) and flowers[i] + k + 1 in record:
                if self.check(record, flowers[i], flowers[i]+k+1):
                    return i + 1
        return -1
    
    def check(self, record, i, j):
        i += 1
        while i < j:
            if i in record:
                return False
            i += 1
        return True
```
